```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("CodTech Task 1 Big Data Analysis") \
    .getOrCreate()
```

```python
df = spark.read.csv("OnlineRetail.csv", header=True, inferSchema=True)
```

```python
df.show(5)
```

```python
print("Total Rows:", df.count())
```

```python
df_clean = df.dropna()
```

```python
df_clean.groupBy("Country") \
    .sum("Quantity") \
    .orderBy("sum(Quantity)", ascending=False) \
    .show(10)
```

```python
df_clean.groupBy("Description") \
    .sum("Quantity") \
    .orderBy("sum(Quantity)", ascending=False) \
    .show(10)
```

INSIGHTS:

1. PySpark efficiently processed a large retail dataset.
2. The United Kingdom has the highest product sales.
3. Certain products have very high sales volumes.
4. PySpark demonstrates scalability through distributed processing.
